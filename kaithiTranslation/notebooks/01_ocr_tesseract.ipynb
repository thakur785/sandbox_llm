```python
#Quick sanity OCR test
import cv2
import numpy as np
import pytesseract

img_path = "sandbox_llm/kaithiTranslation/sampledata/IMG_20251028_185418.jpg"

img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# Increase contrast + remove background noise
img = cv2.adaptiveThreshold(
    img, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    31, 15
)

text = pytesseract.image_to_string(
    img,
    config="--oem 1 --psm 6 -l kat+hin"
)

print(text[:500])
```
